```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("website_traffic.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
df.fillna("Unknown", inplace=True)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.to_csv("cleaned_website_traffic.csv", index=False)
print("Cleaned file saved successfully")
```

```python
total_events = len(df)
print("Total Events:", total_events)
```

```python
unique_users = df['linkid'].nunique()
print("Unique Users:", unique_users)
```

```python
df['event'].value_counts()
```

```python
df['event'].value_counts().plot(kind='bar')
plt.title("Event Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries")
plt.xticks(rotation=45)
plt.show()
```

```python
df['city'].value_counts().head(10)
```

```python
df['city'].value_counts().head(10).plot(kind='bar')
plt.title("Top Cities")
plt.xticks(rotation=45)
plt.show()
```

```python
df['artist'].value_counts().head(10)
```

```python
df['track'].value_counts().head(10).plot(kind='bar')
plt.title("Top Tracks")
plt.xticks(rotation=45)
plt.show()
```

```python
daily_activity = df.groupby('date').size()

daily_activity
```

```python
daily_activity.plot()
plt.title("Daily Activity Trend")
plt.show()
```

```python
print("== FINAL SUMMARY ==")
print("Total Events:", total_events)
print("Unique Users:", unique_users)
print("Top Country:", df['country'].value_counts().idxmax())
print("Top Artist:", df['artist'].value_counts().idxmax())
print("Top Track:", df['track'].value_counts().idxmax())
```
